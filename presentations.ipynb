```python
from playwright.async_api import async_playwright
from dotenv import load_dotenv
import os
import logging

from src.presentations_module.sources.sokratic_source import SokraticSource

load_dotenv()
SOKRATIC_USERNAME = os.getenv("SOKRATIC_USERNAME")
SOKRATIC_PASSWORD = os.getenv("SOKRATIC_PASSWORD")

apw = await async_playwright().start()
logger = logging.getLogger("sokratic_source")

source = SokraticSource(apw, logger)
await source.init_async()

await source.authenticate(login=SOKRATIC_USERNAME, password=SOKRATIC_PASSWORD)

async for response in source.generate_presentation(
    topic="Влияние климата на экосистемы",
    language="kz",
    slides_amount=10,
    audience="Средняя школа",
    author="Кривощеков Артем",
):
    print(response)
```
