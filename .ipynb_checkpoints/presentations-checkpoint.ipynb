```python
import asyncio

from playwright.async_api import async_playwright

from sources.sokratic_source import SokraticSource

apw = await async_playwright().start()

source = SokraticSource(apw)
await source.init_async()

await source.authenticate(login="artiv09@gmail.com", password="Barck1994")

page = await source.generate_presentation(
    topic="Влияние климата на экосистемы",
    language="kz",
    slides_amount=10,
    audience="Средняя школа",
    author="Кривощеков Артем",
)
```

```python
import logging

# Get logger
logger = logging.getLogger("my logger")

# Create a handler
c_handler = logging.StreamHandler()

# link handler to logger
logger.addHandler(c_handler)

# Set logging level to the logger
logger.setLevel(logging.DEBUG) # <-- THIS!
logger.debug('This is a debug message') # WILL WORKlogger.setLevel(logging.DEBUG) # <-- THIS!
```

```python
timeout = 1000 * 60 * 10

async def downloadPresentation(format):
    logger.debug('Check ref window')
    if await page.locator("//div[@role='dialog']//h2[normalize-space(.)='Реферальная программа']").count() > 0:
        logger.debug("Popup window detected, closing")
        await page.locator("//div[@role='dialog']//button[contains(@class, 'top-4')]").click()
    else:
        logger.info('Popup window not detected, continue')
        
    logger.debug('Click download button')
    await page.locator("//button[normalize-space(.)='Скачать']").click()
    
    logger.debug('Click download format button')
    await page.locator("//div[@role='menuitem'][normalize-space(.)='{format}']".format(format=format)
        .click(timeout=timeout)
```

```python
await downloadPresentation('PowerPoint')
```

```python
await downloadPresentation('PDF')
```
